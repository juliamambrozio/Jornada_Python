### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

```python
#!pip install scikit-learn
# !python.exe -m pip install --upgrade pip 

# Entender o desafio
# Importar dados

import pandas 

tabela = pandas.read_csv("clientes.csv")
display(tabela)
```

```python
print(tabela.info())
""" float: números com casas decimais
int: números sem casas decimais
object: string """
```

```python
# Preparar base de dados (pré processamento)
# IA NÃO APRENDE COM OBJECT, SÓ PODE TER OBJECT NO RETORNO DE RESPOSTA
# profissão, mix-credito, comportamento-pagamento
# label enconder = colocar rótulo em cada valor
from sklearn.preprocessing import LabelEncoder # importa só a função que você quer

codificador = LabelEncoder()
tabela["profissao"] = codificador.fit_transform(tabela["profissao"]) # o valor será substituído pela transformação
tabela["mix_credito"] = codificador.fit_transform(tabela["mix_credito"])
tabela["comportamento_pagamento"] = codificador.fit_transform(tabela["comportamento_pagamento"]) 
print(tabela.info())
```

```python
# Quais dados vai utilizar (x)
x = tabela.drop(columns = ["id_cliente", "score_credito"]) # a tabela toda, menos essas duas
# Qual coluna tem que prever (y)
y = tabela["score_credito"]
display(x)


from sklearn.model_selection import train_test_split
# DIVIDE OS DADOS EM 4
# Dados X de treino (para a IA treinar e aprender a fazer)
# Dados Y de teste (para ver se a IA aprendeu)

# Dados X de treino (para a IA treinar e aprender a fazer)
# Dados Y de teste (para ver se a IA aprendeu)

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y) # DIVIDE OS DADOS EM TESTE E TREINO
```

```python
# Criar modelo de IA, calcular nota de crédito (ruim, ok, bom)
# DOIS MODELOS (A forma como aprende e prevê muda para cada modelo)
""" CRIAR IA
importar a IA
criar a IA
treinar a IA """

"""
- Árvore de decisão (faz perguntas para a base de dados e segue caminhos diferentes conforme a resposta (tipo if/else)) 
- KNN (Vizinhos próximos) (Vai separar a base de dados e vai deduzir a resposta de acordo com o valor do elemento 
mais próximo) 
"""
# importar
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
# criar
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()
# treinar
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Escolher o melhor modelo
# Verificar qual acerta mais comparado com o gráfico

previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste) # PREVER
previsao_knn = modelo_knn.predict(x_teste) # PREVER

from sklearn.metrics import accuracy_score # verificar quantos acertou
# prego cada resultado das previsões que ele fez e comparo fazendo a porcentagem de acerto (comparando com o y_teste)
display(accuracy_score(y_teste, previsao_arvoredecisao)) # 82% (O MELHOR)
display(accuracy_score(y_teste, previsao_knn)) # 74% (Para melhorar os dados: mais ou menos colunas, mais dados, 
#mudar modelo)
```

```python
# Usar modelo para fazer previsões
# Importar novos clientes
tabela2 = pandas.read_csv("novos_clientes.csv")
display(tabela2)
# Transformar valores dos dados
tabela2["profissao"] = codificador.fit_transform(tabela2["profissao"]) # o valor será substituído pela transformação
tabela2["mix_credito"] = codificador.fit_transform(tabela2["mix_credito"])
tabela2["comportamento_pagamento"] = codificador.fit_transform(tabela2["comportamento_pagamento"]) 
print(tabela2.info())

# Fazer previsões (usando árvore de decisão)
previsao = modelo_arvoredecisao.predict(tabela2) # passando a tabela toda para prever
display(tabela2)
display(previsao)
```
